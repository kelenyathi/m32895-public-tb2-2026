## `if-else` statements
Boolean logic checked by `if` and `else` statements - **flow control by decision making**.

```python
number = int(input("Enter a number:"))  # input() function is fed to the int() function

if number == 10:
    print(f"{number} is equal to 10")  # String literals
else:
    print(f"{number} is not equal to 10")  # String literals
    
    # Note: formatting of the snippet
    # !! if-else statement can be nested !! This will be important for an in-class exersise.
```

## `if-elif-else` statements
Boolean logic checked by `if`, `elif` and `else` statements - **flow control by decision making**.

```python
a = 200
b = 300
if b > a:
	print("b is greater than a")
elif a == b:
  	print("a and b are equal")
else:
  	print("a is greater than b")
```
